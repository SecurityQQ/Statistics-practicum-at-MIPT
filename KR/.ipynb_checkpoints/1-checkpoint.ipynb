```python
import numpy as np
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
N = 1000
K = 10000
```

```python
samples = np.array([stats.norm.rvs(size=N) for _ in range(K)])
```

```python
mle_estimations = np.array([np.array([np.average(samples[i][:j])
                                      for j in range(1, K)])
                            for i in range(N)])
                   
```

```python
again_est = np.array([np.array([np.min(samples[i][:j])/2 + np.max(samples[i][:j])/2
                                for j in range(1, K)])
                      for i in range(N)])
```

```python
def S(X):
    return np.average(X**2) - (np.average(X))**2
```

```python
s2_1 = np.array([S(mle_estimations[i]) for i in range(len(mle_estimations))])
```

```python
s2_2 = np.array([S(again_est[i]) for i in range(len(again_est))])
```

```python
def draw_plot():
    plt.figure(figsize=(15, 9))
    plt.plot(s2_1, label='with MLE')
    plt.plot(s2_2, label='(min+max) / 2')
    plt.legend(loc='best')
    plt.show()
```

```python
draw_plot()
```

```python
draw_plot(0.001)
```
