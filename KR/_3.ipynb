```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('marks.csv', header = None)
```

```python
data.head()
```

```python
len(data)
```

```python
alphas = []
for i in range(1, 14):
    alphas.append(data[i].value_counts() / len(data[i].value_counts()))
```

```python
for i in range(len(data)):
    betas = data.ix[i,1:].value_counts()
    for j in range(1, 14): 
        if np.isnan(data.ix[i,j]):
            gammas = alphas[j - 1]
            for index in betas.index:
                if index in alphas[j - 1].index:
                    gammas[index] += betas[index]
            expect = 0
            for index in gammas.index:
                expect += gammas[index] * index / sum(gammas)
            data.ix[i,j] = expect
                
```

```python
data['grade'] = [np.sum(data.ix[i,1:]) / 13 / 3 * 10 for i in range(len(data))]
```

```python
def get_grade(x):
    if x < 3:
        return 2
    elif x < 5:
        return 3
    elif x < 8:
        return 4
    else:
        return 5
```

```python
data['grade_5'] = [get_grade(data['grade'][i]) for i in range(len(data))]
```
