```python
import numpy as np
import scipy.stats as sps
import matplotlib.pylab as plt
%matplotlib inline
```

```python
N = 1000
K = 10000
sample = sps.norm().rvs(size = N * K)
```

```python
sample = sample.reshape(N, K)
```

```python
est_1 = np.array([np.mean(sample[:i + 1], axis=0) for i in range(N)])
```

```python
est_2 = np.array([(np.min(sample[:i + 1], axis=0) + 
                np.max(sample[:i + 1], axis=0)) / 2.0 for i in range(N)])
```

```python
sigma_1 = np.var(est_1, axis=1)
```

```python
sigma_2 = np.var(est_2, axis=1)
```

```python
def draw_plot(yl):
    plt.figure(figsize=(15, 9))
    plt.plot(range(N), sigma_1, label='MLE')
    plt.plot(range(N), sigma_2, label='min/2 + max/2')
    plt.legend(loc='best')
    plt.ylim(0, yl)
    plt.show()
```

```python
draw_plot(0.3)
```
